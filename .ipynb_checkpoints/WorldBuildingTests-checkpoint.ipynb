## Population Demographics Simulation

```python
class Race:

    def __init__(self,world,name,parent = "Originator",color = 0,type = "Human", movementspd = 10,t0 = 0, growthrate = 0.4,deathrate = 1,renewabledepletion = 0.01,nonrenewabledepletion = 0.001,startRain = 0.7,crowding = 0.1,idealTemp = 0.75):

        self.Name = name
        self.Type = type
        self.Parent = parent

        PossibleStarts = np.where((world.Elevation > world.oLevel)*(world.RainFall > startRain))
        if len(PossibleStarts[0]):
            pk = np.random.randint(len(PossibleStarts[0]))
        else:
            PossibleStarts = np.where((world.Elevation > world.oLevel))
            pk = np.random.randint(len(PossibleStarts[0]))
        thept = PossibleStarts[0][pk]*world.GlobeGrid[0].shape[0] + PossibleStarts[1][pk]

        InitialPop = np.zeros(len(world.gridindices))
        InitialPop[thept] = 1
        self.InitialDistribution = InitialPop.reshape(world.GlobeGrid[0].shape)
        self.Population = self.InitialDistribution.copy()

        grs = growthrate*np.ones(len(InitialPop))
        grs[world.OceanIndicator]  = 0
        self.growthrates = grs.reshape(world.GlobeGrid[0].shape)

        drs = deathrate*np.ones(len(InitialPop))
        drs[world.OceanIndicator]  = 0
        self.deathrates = drs.reshape(world.GlobeGrid[0].shape)

        rds = renewabledepletion*np.ones(len(world.InitialRenew.flatten()))
        rds[world.OceanIndicator]  = 0
        self.RenewDeplete = rds

        nrds = nonrenewabledepletion*np.ones(len(world.InitialRenew.flatten()))
        nrds[world.OceanIndicator]  = 0
        self.NonRenewDeplete = nrds

        self.Movement = movementspd
        #if color == 'r':
        if hasattr(color, "__len__"):
            if len(color) == 3:
                self.BaseChromosome = color
            else:
                self.BaseChromosome = np.array(cm.nipy_spectral(np.random.rand())[:3])#np.random.rand(3)
        else:
            self.BaseChromosome = np.array(cm.nipy_spectral(np.random.rand())[:3])#np.random.rand(3)


        self.Chromosomes = self.BaseChromosome*self.InitialDistribution.astype(bool)[:,:,None]

        self.Cities = np.zeros_like(self.Population)

        self.History = {"Population":[self.InitialDistribution.copy()], "Genetics":[self.Chromosomes.copy()],"Cities":[self.Cities], "Time":[t0]}

        self.IdealTemp = idealTemp
        self.Crowding = crowding
```

```python
def desire(pop,avgTemp,R,R0,Crowding,GT,River,Ocean):
    return np.exp(-(pop-Crowding)**2)*(R/R0)*np.exp(-(avgTemp-GT)**2)*Ocean.astype(float)*(1 + 0.3*River.astype(float))
#     if Ocean:
#         return 0.0
#     else:
#         crowds = np.exp(-(pop-Crowding)**2)
#         resource = R/R0
#         Temp = np.exp(-(AvgTemp-GT)**2)
#         if River:
#             return (1.3)*crowds*resource*Temp
#         else:
            return crowds*resource*Temp
```

```python
def envy(world,raceList):
    
    totalPop = sum([rc.Population for rc in raceList])
    totalRes = world.Renewables + world.NonRenewables
    
    popMeans,popDevs = compare(totalPop)
    resMeans,resDevs = compare(totalRes)
    
    popMeans = popMeans/np.max(popMeans)
    resMeans = resMeans/np.max(resMeans)
    
    scldpopDevs = np.arctan(popDevs)/np.pi + 0.5
    scldresDevs = np.arctan(resDevs)/np.pi + 0.5
    
    popEnv = popMeans*scldpopDevs
    resEnv = resMeans*scldresDevs
    
    return np.divide(popEnv,resEnv, out = np.zeros_like(popEnv), where = resEnv>0)
    
    
def compare(data):
    N,M = data.shape
    means = np.array([[np.mean(getSurrounding((i,j),data)) for j in range(M)] for i in range(N)])
    stddeviations = np.array([[np.std(getSurrounding((i,j),data)) for j in range(M)] for i in range(N)])
    absDiffs = data-means
    signedDeviation = np.divide(absDiffs,variances, out = np.zeros_like(absDiffs), where = stddeviations>0)
    return means,signedDeviation

def LocalGeneticVariation(pt,raceList):
    surroundingpops =[getSurrounding(pt,rc.Population) for rc in raceList]
    totspops = sum(surroundingpops)
    surroundingpops = [np.divide(sp,totspops,out = np.zeros_like(sp),where = totspos>0) for sp in surroundingpops]
    surrChroms = sum([getSurrounding(pt,raceList[i].Chromosomes)*surroundingpops[i][:,:,None] for i in range(len(raceList))])
    rtsmsqdiffs = (np.sum((surrChroms[1,1,:] - surrChroms)**2, axis = 2))**(0.5)
    return rtsmsqdiffs
```

```python
def lamGrowths(p,r,n,kap):
    return p*r*n*kap

def lamDeaths(p,gam):
    return gam*p

def lamLocalMovement(p,world,desirabilities,all_dists,mvmnt = 1):
    N  = world.GlobeGrid[0].shape[0]
    M = world.GlobeGrid[0].shape[1]
#     gridarr = [(i,j) for i in range(world.GlobeGrid[0].shape[0]) for j in range(world.GlobeGrid[0].shape[1])]
    
    all_locMoveLamsOut = np.array([[getSurrounding((i,j),desirabilities) for j in range(M)] for i in range(N)])
    all_locMoveLamsOut = all_locMoveLamsOut*p[:,:,None,None]
    all_locMoveLamsOut = np.divide(all_locMoveLamsOut,all_dists, out = all_locMoveLamsOut/mvmnt,where=all_dists>0)
    
    
    return all_locMoveLamsOut
    

def lamRenews(r,R0, alpha):
    return alpha*(np.array(R0)-np.array(r))
```

```python
def GrowAll(t,y,params):
    world,racelist = params
    nmspots = len(world.GlobeGrid[0].flatten())
    
    sepVars = y.reshape(len(racelist) + 2,nmspots)
    
    alldots = np.zeros((len(racelist) + 2,nmspots))
    
    for rc in range(len(racelist)):
        grth = Growth(t,sepVars[rc],sepVars[-2],sepVars[-1],[world,racelist[rc]])
        alldots[rc] = grth[0]
        alldots[-2] = alldots[-2] + grth[1]
        alldots[-1] = alldots[-1] + grth[2]
        
    return alldots.flatten()
        
    

def Growth(t,population,Renews,NonRenews,params):
    
    world,race = params
    
    growthrates = race.growthrates.flatten()
    deathrates = race.deathrates.flatten()
    mmnt = race.Movement
    rdeplete = race.RenewDeplete.flatten()
    nrdeplete = race.NonRenewDeplete.flatten()
        
    renewrate = world.renewrate.flatten()
    R0s = world.InitialRenew.flatten()
        
    
     ###Get our lambdas!
    gLams = lamGrowths(population,Renews,NonRenews,growthrates)
    dLams = lamDeaths(population,deathrates)
    renewLams = lamRenews(Renews,R0s,renewrate)
    
    xDot = (gLams - dLams)
    RenewDot = (renewLams - rdeplete*gLams)
    NonRenewDot =(-nrdeplete*gLams)
    
    ydot = np.array([xDot,RenewDot,NonRenewDot])
    
    
    return ydot

def getDestination(i,j,probs,gridshape):
    N,M = gridshape
    neighborar = [(-1, -1), (-1, 0), (-1, 1), (0, -1), (0,0) , (0, 1), (1, -1), (1, 0), (1, 1)]
    chc = neighborar[np.random.choice(9,p = probs)]
    return ((i + chc[0]) % N, max(0,min((j+chc[1]),M-1)))

def moveAround(world,raceset):
         
    gridpoints = world.gridindices
    RiverHere = world.RiverIndicator.reshape(world.GlobeGrid[0].shape)
    OceanHere = world.OceanIndicator.reshape(world.GlobeGrid[0].shape)
    disToNeighbors = world.NeighborDistances
    R0s = world.InitialRenew
    disfun = world.GCDistance
    coast = world.CoastIndicator
    
#     chroms = [race.Chromosomes for race in raceset]
    
    numpts = len(raceset[0].Population.flatten())
#     populations = [race.Population for race in raceset]
    totalPop = np.sum(populations, axis = 0)
        
    neighborar = [(-1, -1), (-1, 0), (-1, 1), (0, -1), (0,0) , (0, 1), (1, -1), (1, 0), (1, 1)]
    
    rmean =(np.sum(world.Renewables) + np.sum(world.NonRenewables))/(len(world.Renewables.flatten()) + len(world.NonRenewables.flatten()))
    crowded = 0.1
    idealTemp = 0.75
    
    AvgTemps = (world.Temps[0] + world.Temps[1])/2
    
#     Alldesirabilities = [desire(totalPop,AvgTemps,world.Renewables + world.NonRenewables,rmean,race.Crowding,race.IdealTemp,OceanHere,RiverHere) for race in raceset]
    
    for race in raceset:
        
        Alldesirabilities = desire(totalPop,AvgTemps,world.Renewables + world.NonRenewables,rmean,race.Crowding,race.IdealTemp,OceanHere,RiverHere)
        
         ###Get our lambdas!
        moveLamsOut = lamLocalMovement(population,world,Alldesirabilities,disToNeighbors,mvmnt = race.Movement)
        moveLamsOut = np.maximum(moveLamsOut,np.zeros_like(moveLamsOut))
        tots = np.sum(np.sum(moveLamsOut,axis = 3),axis = 2)
        probs = np.divide(moveLamsOut,tots[:,:,None,None],out=np.zeros_like(moveLamsOut),where=tots[:,:,None,None]>0)
        probs[np.where(tots == 0)] = np.diag([0,1,0])
        all_choices =[[getDestination(i,j,probs[i,j].flatten(),world.GlobeGrid[0].shape) for j in range(world.GlobeGrid[0].shape[1])] for i in range(world.GlobeGrid[0].shape[0])]


        prevpopu = race.Population#population.reshape(world.GlobeGrid[0].shape)
        newpop = prevpopu.copy()
        
        chng = np.random.rand()*0.01*prevpopu

        for i in range(world.GlobeGrid[0].shape[0]):
            for j in range(world.GlobeGrid[0].shape[1]):
                chng = np.random.rand()*0.01*prevpopu[i,j]
                if chng:
                    chc = neighborar[all_choices[i][j]]
                    newgridpt =((i + chc[0]) % world.GlobeGrid[0].shape[0], min(max(j+chc[1],0),world.GlobeGrid[0].shape[1]-1))
                    newpop[i,j] = newpop[i,j] - chng
                    newpop[newgridpt] = newpop[newgridpt] + chng
                    if prevpopu[newgridpt] == 0:
                        chrom[newgridpt] = chrom[i,j]

        ###### Add some ocean crossing.
        coastpops = prevpopu*coast.astype(float)
        totcoast = np.sum(coastpops)
        freeCoast = coast*np.invert(prevpopu.astype(bool))
        if totcoast.round(8) and np.sum(freeCoast):

            cpts = np.where(world.CoastIndicator)
            surroundings = (sum([np.sum(getSurrounding((cpts[0][i],cpts[1][i]),prevpopu)) - prevpopu[cpts[0][i],cpts[1][i]] for i in range(len(cpts[0]))]))/8

            if np.random.exponential(1/totcoast) < surroundings:

                coastChc = np.random.choice(range(len(coastpops.flatten())), p = coastpops.flatten()/totcoast)
                leavingInd = np.unravel_index(coastChc,coastpops.shape)
                leavp =(world.GlobeGrid[0][leavingInd],world.GlobeGrid[1][leavingInd])
                leavingPop = 0.01*np.random.rand()*prevpopu[leavingInd]
                ## Randomly choose an open coastal spot
                distsFrom = np.array([disfun(leavp,(world.GlobeGrid[0][i,j],world.GlobeGrid[1][i,j])) if freeCoast[i,j] else 0  for i in range(world.GlobeGrid[0].shape[0]) for j in range(world.GlobeGrid[0].shape[1])])
                distsFrom = np.divide(np.ones_like(distsFrom),distsFrom,out = np.zeros_like(distsFrom), where=distsFrom.round(8)>0)
                newChc = np.random.choice(range(len(freeCoast.flatten())), p =np.array(distsFrom)/sum(distsFrom))
                newInd = np.unravel_index(newChc,prevpopu.shape)

                newpop[leavingInd] = newpop[leavingInd] - leavingPop
                newpop[newInd] = newpop[newInd] + leavingPop
                chrom[newInd] = chrom[leavingInd]


    return newpop, chrom

def weightedSurroundingChromosome(pt,race):
        
    pop = getSurrounding(pt,race.Population)
    chrom = getSurrounding(pt,race.Chromosomes)
        
    weightMat = np.array([[1,1,1],[1,5,1],[1,1,1]])*pop
    prod = chrom*weightMat[:,:,None]
    if np.sum(weightMat):
        return np.sum(np.sum(prod/np.sum(weightMat),axis = 0),axis = 0)
    else:
        return chrom[1,1]

def evolve(raceset,mutationrate = 0.005):
    gridShape = race.Population.shape
    racesBases = [np.array([[weightedSurroundingChromosome((i,j),rc) for j in range(gridShape[1])] for i in range(gridShape[0])]) for rc in allraces]
    racesBases = dict([(allraces[i].Name,racesBases[i]) for i in range(len(allraces))])
    newChroms = racesBases[race.Name]
    
    popsum = sum([rc.Population for rc in allraces])
    population_proportions = dict([(rc.Name,np.divide(rc.Population,popsum,out = np.zeros_like(popsum),where=popsum>0)) for rc in allraces])
    
    diffa = [racesBases[rc.Name] - racesBases[race.Name] for rc in allraces]
    coeffs = [0.2*np.random.rand(*gridShape)*population_proportions[rc.Name] if rc.Type == race.Type else 0.1*np.random.rand(*gridShape)*population_proportions[rc.Name] for rc in allraces]
    newChroms = newChroms + sum([diffa[i]*coeffs[i][:,:,None] for i in range(len(allraces))])
    
    ##add a little random wiggle
    randVar = (np.random.rand(*gridShape,3) - 0.5)*mutationrate
    
    newChroms = newChroms + randVar*race.Population.astype(bool)[:,:,None]
    
    newChroms = np.maximum(np.minimum(newChroms,np.ones_like(race.Chromosomes)), np.zeros_like(race.Chromosomes))    
    
    return newChroms*race.Population.astype(bool)[:,:,None]
    
```

```python
def getSurrounding(j,D):
    N,M = D.shape[:2]
    return np.array([[D[((j[0] + l) % N, max(0,min((j[1]+m),M-1)))] for m in [-1,0,1]] for l in [-1,0,1]])
```

```python
def randomCities(Population,prevCities,peoplePerCity):
    ##total global population
    poptot = np.sum(Population)
    cities = np.zeros_like(Population)
    if poptot:
        num_cities = int(poptot/peoplePerCity)
        prvBoost = np.sum(Population>0)
        adjPop = Population + 2*(poptot*prevCities)#/prvBoost
        newTot = np.sum(adjPop)
        scldPop = adjPop/newTot
        for i in range(num_cities):
            cityin = np.unravel_index(np.random.choice(np.arange(len(Population.flatten())), p = scldPop.flatten()),Population.shape)
            cities[cityin] = cities[cityin] + 1
            adjPop[cityin] = np.maximum(adjPop[cityin] - peoplePerCity*poptot/prvBoost,0)
            newTot = np.sum(adjPop)
            if newTot:
                scldPop = adjPop/newTot
            else:
                break
    return cities
```

```python
def GetNewBase(race,world,current_bases):
    dists = [np.linalg.norm(race.Chromosomes - baseC, axis = 2) for baseC in current_bases]
    newRaceMask = np.prod([d>0.6 for d in dists],axis = 0)*world.LandIndicator*race.Population.astype(bool)
    if np.sum(newRaceMask):
        n1 = np.random.choice(len(race.Population.flatten()),p = newRaceMask.flatten()/np.sum(newRaceMask))
        newBaseChromosome = race.Chromosomes[np.unravel_index(n1, race.Population.shape)]
        return newBaseChromosome, True
    else:
        return None, False
    
    
def GetRaceMask(race,base):
    dists = np.linalg.norm(race.Chromosomes - base, axis = 2)
    return dists < 0.1
    
    
def splitRace(world,race):
    spawnedRaces = [(race.BaseChromosome,GetRaceMask(race,race.BaseChromosome))]
    stillmore = True
    while stillmore:
        newBase,stillmore = GetNewBase(race,world,[ra[0] for ra in spawnedRaces])
        if stillmore:
            spawnedRaces += [(newBase,GetRaceMask(race,newBase))]
    return spawnedRaces

def newRaces(world,racelist,t):
    allRaces = [splitRace(world,rc) for rc in racelist]
    newRacelist = racelist.copy()
    racebases = [rc.BaseChromosome for rc in racelist]
    for i in range(len(racelist)):
        for newR in allRaces[i][1:]:
            disFromExisting = np.linalg.norm(newR[0]-np.array(racebases), axis = 1) < 0.3
            if any(disFromExisting):
                #merge with the one it's close to
                raceinto = newRacelist[np.where(disFromExisting)[0][0]]
                newRacelist[np.where(disFromExisting)[0][0]].Population  = raceinto.Population + racelist[i].Population*newR[1]
                newRacelist[np.where(disFromExisting)[0][0]].Chromosomes[newR[1]] = 0.5*(raceinto.Chromosomes[newR[1]] + racelist[i].Chromosomes[newR[1]])
                for nm in racelist[i].Name.split('-'):      
                    if nm not in raceinto.Parent:
                        newRacelist[np.where(disFromExisting)[0][0]].Parent = raceinto.Parent + '-' + nm
                for tp in racelist[i].Type.split('-'):
                    if tp not in raceinto.Type:
                        newRacelist[np.where(disFromExisting)[0][0]].Type = newRacelist[np.where(disFromExisting)[0][0]].Type + tp
            else:
                newTag = ''.join([np.random.choice(list('0123456789abcdefghijklmnopqrstuvwxyz')) for n in range(6)])
                newRace = Race(world,newTag, parent = racelist[i].Name,t0 = t,color = newR[0])
                newRace.InitialDistribution = racelist[i].Population*newR[1]
                newRace.Population = newRace.InitialDistribution
                newRace.Chromosomes = np.zeros_like(racelist[i].Chromosomes)
                newRace.Chromosomes[newR[1]] = racelist[i].Chromosomes[newR[1]]
                newRace.Type = racelist[i].Type
                newRace.growthrates = racelist[i].growthrates
                newRace.deathrates = racelist[i].deathrates
                newRace.RenewDeplete = racelist[i].RenewDeplete
                newRace.NonRenewDeplete = racelist[i].NonRenewDeplete
                newRace.History = {"Population":[newRace.InitialDistribution.copy()], "Genetics":[newRace.Chromosomes.copy()],"Cities":[newRace.Cities], "Time":[t]}
                newRacelist += [newRace]
                racebases += [newRace.BaseChromosome]
    return newRacelist
    
```

```python
humans = Race(World1,"ProtoHumans",type = "Human", parent = "God", color = tuple(np.array([0,0.5,0.5])))
orcs = Race(World1,"ProtoOrcs",type = "Orcs",parent = "God", color = tuple(np.array([210,105,30])/255))
elves = Race(World1,"ProroElves",type = "Elven",parent = "God", color = tuple(np.array([255,20,147])/255),growthrate = 0.2,deathrate = 0.5)
raceOrder = [humans,orcs,elves]
```

```python
resources = {"Renewables":[World1.InitialRenew],"NonRenewables":[World1.InitialNonRenew],"Time":[0]}

dt = 0.01
endt = 11

allRaces = np.array([race.Population.flatten() for race in raceOrder]).flatten()
fullState = np.concatenate([allRaces,World1.InitialRenew.flatten(),World1.InitialNonRenew.flatten()])

simWorld = ode(GrowAll)
simWorld.set_initial_value(fullState, 0).set_f_params([World1,raceOrder])


while resources["Time"][-1]<endt:
    #we'll do an alternating update scheme. Grow.
    fullState = simWorld.integrate(simWorld.t + dt).reshape(len(raceOrder) + 2, len(World1.GlobeGrid[0].flatten()))
    #then random instantaneus movement to neighbors.
    
    World1.Renewables = fullState[-2].reshape(World1.GlobeGrid[0].shape)
    resources["Renewables"] += [World1.Renewables]
    World1.NonRenewables = fullState[-1].reshape(World1.GlobeGrid[0].shape)
    resources["NonRenewables"] += [World1.NonRenewables]
    
    for rci in range(len(raceOrder)):
        rc = raceOrder[rci]
        rc.Population = fullState[rci].reshape(World1.GlobeGrid[0].shape)
        rc.Population,rc.Chromosomes = moveAround(World1,rc)
        rc.Chromosomes = evolve(rc,raceOrder)
        rc.Cities = randomCities(rc.Population,rc.Cities,1000)
        rc.History["Genetics"] += [rc.Chromosomes]
        rc.History["Population"] += [rc.Population]
        rc.History["Cities"] += [rc.Cities]
        rc.History["Time"] += [resources["Time"][-1] + simWorld.t]

    resources["Time"] += [resources["Time"][-1] + simWorld.t]
    
#     if not (len(resources["Time"]) % 500):
#         raceOrder = newRaces(World1,raceOrder,resources["Time"][-1])
#         print(resources["Time"][-1],len(raceOrder))
    
    allRaces = np.array([race.Population.flatten() for race in raceOrder]).flatten()
    fullState = np.concatenate([allRaces,fullState[-2],fullState[-1]])
    simWorld.set_initial_value(fullState,0).set_f_params([World1,raceOrder])
```

```python
def GetNewBase(race,world,current_bases):
    dists = [np.linalg.norm(race.Chromosomes - baseC, axis = 2) for baseC in current_bases]
    newRaceMask = np.prod([d>0.1 for d in dists],axis = 0)*world.LandIndicator
    if np.sum(newRaceMask):
        n1 = np.random.choice(len(race.Population.flatten()),p = newRaceMask.flatten()/np.sum(newRaceMask))
        newBaseChromosome = race.Chromosomes[np.unravel_index(n1, race.Population.shape)]
        return newBaseChromosome, True
    else:
        return None, False
    
    
def GetRaceMask(race,base):
    dists = np.linalg.norm(race.Chromosomes - base, axis = 2)
    return dists < 0.1
    
    
def splitRace(world,race):
    bases = [race.BaseChromosome]
    racemasks = [GetRaceMask(race,race.BaseChromosome)]
    stillmore = True
    while stillmore:
        newBase,stillmore = GetNewBase(race,world,bases)
        if stillmore:
            bases += [newBase]
            racemasks += [GetRaceMask(race,newBase)]
    return bases,racemasks
```

```python
def cityTimes(pt,citieslist,dt):
    city_indicator = [bool(ci[pt]) for ci in citieslist]
    uncounted = city_indicator.copy()
    city_times = []
    while any(uncounted):
        i1 = np.where(uncounted)[0][0]
        t1 = i1*dt
        t2 = t1
        while uncounted[i1]:
            uncounted[i1] = False
            t2 += dt
            i1 += 1
            if i1>=len(city_indicator):
                break
        city_times += [(t1,t2)]
            
    return city_times
```
